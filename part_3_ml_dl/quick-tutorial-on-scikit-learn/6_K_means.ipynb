```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn import metrics

%matplotlib inline
```

```python
blobs = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.5, random_state=101)
```

```python
blobs[0][:10]
```

```python
blobs[1][:10]
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(blobs[0][:,0], blobs[0][:,1], hue=blobs[1], palette='plasma_r')
```

```python
k_means = KMeans(n_clusters=4)
```

```python
k_means.fit(blobs[0])
```

We can get the coordinations of the cluster centroids by:

```python
k_means.cluster_centers_
```

For see the clusters assigned to each data point, we can do:

```python
k_means.labels_[:10]
```

```python
fig, (ax_1, ax_2) = plt.subplots(1, 2, figsize=(10, 5))
sns.despine(left=True)

ax_1.set_title('Ground Truth')
sns.scatterplot(blobs[0][:,0], blobs[0][:,1], hue=blobs[1], palette='plasma_r', ax=ax_1)

ax_2.set_title('K Means')
sns.scatterplot(blobs[0][:,0], blobs[0][:,1], k_means.labels_, palette='plasma_r', ax=ax_2)
```
