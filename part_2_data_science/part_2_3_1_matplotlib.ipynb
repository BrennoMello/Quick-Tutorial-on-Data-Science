## Introduction

```python
import matplotlib.pyplot as plt
import numpy as np
import pylab
```

```python
%matplotlib inline
pylab.rcParams['figure.figsize'] = (4.0, 3.0)
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 55, 42, 45, 38, 44, 51]
```

We may draw a plot as follows for variables x and y. We may use `xlabel` and `ylabel` to set labels for axes.

```python
plt.title('Weather')

plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(x, y, color='red')
```

---

### Formatting Arguments in Plot

```python
plt.plot(x, y, color='blue', linewidth=2, linestyle='dotted')
```

```python
plt.plot(x, y, 'ro', markersize=10)
```

```python
plt.plot(x, y, 'g*-.', markersize=10) 
```

Please note that the order of characters does not matter, e.g., `plt.plot(x, y, '--g+')` results in the same plot.

---

### Markers

**Marker**: This argument is used to set the marker shape. In the example below, `D` stands for `Diamond`.

```python
plt.plot(x, y, color='red', marker='D', linestyle='')
```

One may combine parameters `color` and `marker` as below to draw markers:

```python
plt.plot(x, y, 'rD')
```

**Markersize**: The parameter `markersize` defines the size of markers. For example, we set the marker size to `10` in the following example.

```python
plt.plot(x, y, color='red', marker='D', linestyle='', markersize=10)
```

**Transparency**: The argument `alpha` is used to set the transparency rate. We set the transparency rate to `0.5` in the plot below.

```python
plt.plot(x, y, color='red', marker='D', linestyle='', markersize=10, alpha=0.5)
```

---

### Axes and Legend

I show how to add axes labels and a legend to a plot below. For axes labels, we should use `xlabel` to set the label for axis-x, and `ylabel` for axis-y. 

One needs to call the `legend` function for having a lengend on the plot. The function accepts `loc`, `shadow`, and `fontsize` as arguments.

We may add grids to the plot by calling the `grid` function.

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

plt.legend(loc="upper right", shadow=True, fontsize="large")  

plt.grid()
```

---

### Plot Appearance: Fine-grained

```python
fig = plt.figure(figsize=(4, 4))

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, 
        color='blue', 
        linewidth=3, # lw works as well
        alpha=0.8, 
        linestyle="-.", # ls works as well 
        marker='*', 
        markersize=20, 
        markerfacecolor='yellow', 
        markeredgewidth=2, 
        markeredgecolor='red')

ax.set_xlim([0, 6])
ax.set_ylim([0, 30])
```

## Charts

### Bar Charts

```python
company = ["GOOGL", "AMZN", "MSFT", "FB"]
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
pos = np.arange(len(company))
```

#### Vertical Bar Charts

For plotting bar charts vertically, we first need to set the ticks using `xticks`, and then apply `bar` function.

```python
plt.title("US Tech Stocks")

plt.xticks(pos, company)
plt.ylabel("Revenue (bln)")

plt.bar(pos, revenue, label="Revenue")
plt.bar(pos, profit, label="Profit")

plt.legend()
```

We may separate bars for each entry by changing the positions, e.g., adding or subtracting `0.2`.

```python
plt.title("US Tech Stocks")

plt.xticks(pos, company)
plt.ylabel("Revenue (bln)")

plt.bar(pos - 0.2, revenue, width=0.4, label="Revenue")
plt.bar(pos + 0.2, profit, width=0.4, label="Profit")

plt.legend()
```

#### Horizontal Bar Charts

For plotting bar charts horizontally, we first need to set the ticks using `yticks`, and then apply `barh` function.

```python
plt.title("US Tech Stocks")

plt.xlabel("Revenue (bln)")
plt.yticks(pos, company)

plt.barh(pos - 0.2, revenue, height=0.4, label="Revenue")
plt.barh(pos + 0.2, profit, height=0.4, label="Profit")

plt.legend()
```

### Histograms

```python
blood_suger = [113, 85, 90,150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

The most important argument that should be set for historgrams is `bins`, it could be set by a number or a list as in the following examples. In the former one, we set the number of bins to `4`, and we observe four bars in the histogram. In the latter one, we use a list to define the chunks of bins.

The parameter `rwidth` defines the width of bars. And so, we can disjoint them.

```python
plt.hist(blood_suger, bins=4, rwidth=0.8)
```

```python
plt.hist(blood_suger, bins=[80, 100, 125, 150], rwidth=0.8, color='g')
```

#### Playing with Parameters

```python
plt.hist(blood_suger, 
         bins=[80, 100, 125, 200], 
         rwidth=0.8, color='g', 
         histtype='step')
```

```python
blood_suger_man = [113, 85, 90,150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_suger_woman = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.title('Blood Sugar Analysis')

plt.xlabel('Sugar Range')
plt.ylabel('Total No. of Patients')

plt.hist([blood_suger_man, blood_suger_woman], 
         bins=[80, 100, 125, 150], 
         rwidth=0.8, color=['b', 'r'], 
         label=['men', 'women'])

plt.legend()
```

```python
plt.hist([blood_suger_man, blood_suger_woman], 
         bins=[80, 100, 125, 150], 
         rwidth=0.8, color=['b', 'r'], 
         label=['men', 'women'], 
         orientation="horizontal")
```

### Pie Charts

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Expenses"]
```

To plot a pie chart, we need to set to parameters `x` and `labels`.

```python
plt.pie(x=exp_vals, labels=exp_labels)
plt.show()
```

We may set axis to `equal` for having a better looking pie chart as follows:

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

One may increase the size of plot by the `radius` parameters. Moreover, the `autopct` parameter may be used to represent the precentage for each pie chunk. We may drag a chunk out by the `explode` option. The `shadow` let use lay a shadow under the pie chart. You may consider following example for those parameters.

```python
plt.axis("equal")

plt.pie(exp_vals, labels=exp_labels,radius=1.5, autopct='%.2f%%', 
        explode=[0, 0.2, 0, 0, 0], shadow=True)

plt.show()
```

---

## Saving a Plot

One may save a plot using the `savefig` function.

```python
plt.axis("equal")

plt.pie(exp_vals, labels=exp_labels,radius=1.5, autopct='%.2f%%', 
        explode=[0, 0.2, 0, 0, 0], shadow=True)

plt.savefig('piechart.png', bbox_inches='tight', pad_inches=1, transparent=True)
plt.savefig('piechart.pdf', bbox_inches='tight', pad_inches=1, transparent=True)
```

## Multiple Plots

One may have multiple plots using the `subplot` function or by a `figure` object.

```python
pylab.rcParams['figure.figsize'] = (5.0, 4.0)

x = np.linspace(0, 5, 11)
y = x ** 2
```

### Solution 1: subplot

We need three parameters for this function. They are:
* The total number of rows
* The total number of columns
* The plot number

**NB.** subplot(#rows, #columns, plot_no.)

Please consider following exmaples.

**Example 1: Basic**

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')

plt.tight_layout()
```

**Example 2: Intermediate**

In this example, we have 2 rows and 3 columns. Therefore, we can have 6 plots in total, as
[1, 2, 3]
[4, 5, 6].

```python
plt.subplot(2, 3, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 3, 2)
plt.plot(y, x, 'b')

plt.subplot(2, 3, 3)
plt.plot(y, y ** 2, 'g')

plt.subplot(2, 3, 5)
plt.plot(x, x ** 2, 'y')

plt.tight_layout()
```

**Example 3: Advanced**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].set_title('First Plot')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].plot(x, y, 'b')

axes[1].set_title('Second Plot')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].plot(y, x, 'r')

plt.tight_layout()
```

### Solution 2: figure

We may draw (multiple) plots using the `add_axes` function as below.

**Example 1: Basic**

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.set_title('Example')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.plot(x, y)
```

**Example 2: Overlapping Plots**

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # left, bottom, width, height
axes_2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes_1.set_title('Outter Plot')
axes_1.set_xlabel('x')
axes_1.set_ylabel('y')
axes_1.plot(x, y)

axes_2.set_title('Inner Plot')
axes_2.set_xlabel('x')
axes_2.set_ylabel('y')
axes_2.plot(y, x, 'r')
```
